```python
import pandas as pd
data = pd.read_csv("insurance_pre.csv")
data
```

```python
data = pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent = data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = data[['charges']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(independent,dependent,test_size=0.3,random_state=0)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
y_train = sc.fit_transform(y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
re = grid.cv_results_ 
print("The R_score value of the best parameter is {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```
